```python
# 5. which mode of call is giving you more result.... cellular or unknown

df1.groupby(['contact']).size()
```

```python
# 6. how many enterpeures do we have in this list  

df1[df1['job'] == 'entrepreneur']
```

```python
len(df1[df1['job'] == 'entrepreneur'])
```

```python
# 7. how many customers do we have with negative balance 

df1.groupby(df1['balance'].apply(lambda x: 'negative' if x < -0.1 else 'positive' if x > 0.1 else 'neutral'))['balance'].count()
```

```python
# another way

len(df1[df1['balance'] < 0])
```
